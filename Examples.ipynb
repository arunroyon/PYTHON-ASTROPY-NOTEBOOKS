# Data Types

```python
# Python Data Types Example

# Integer
integer_variable = 42

# Float
float_variable = 3.14

# String
string_variable = "Hello, Python!"

# List
list_variable = [1, 2, 3, "four", 5.0]

# Tuple
tuple_variable = (10, 20, "thirty", 40.0)

# Set
set_variable = {1, 8, 2, 7, 3, 3, 6, 7, 4, 5}

# Dictionary
dictionary_variable = {'name': 'John', 'age': 25, 'city': 'New York'}

# Displaying the variables
print("Integer:", integer_variable,"  ", type(integer_variable))
print("\n")
print("Float:", float_variable, type(float_variable))
print("\n")
print("String:", string_variable, type(string_variable))
print("\n")
print("List:", list_variable, type(list_variable))
print("\n")
print("Tuple:", tuple_variable, type(tuple_variable))
print("\n")
print("Set:", set_variable, type(set_variable))
print("\n")
print("Dictionary:", dictionary_variable, type(dictionary_variable))
```

# Arithmetic Operators:


```python
# Addition
result_addition = 5 + 3
print("Addition:", result_addition)

# Subtraction
result_subtraction = 10 - 4
print("Subtraction:", result_subtraction)

# Multiplication
result_multiplication = 6 * 7
print("Multiplication:", result_multiplication)

# Division
result_division = 15 / 3
print("Division:", result_division)

# Modulus (remainder)
result_modulus = 17 % 4
print("Modulus:", result_modulus)

# Exponentiation
result_exponentiation = 2 ** 3
print("Exponentiation:", result_exponentiation)
```

# Comparison Operators:

```python
# Equal to
result_equal = (5 == 5)
print("Equal:", result_equal)

# Not equal to
result_not_equal = (10 != 5)
print("Not Equal:", result_not_equal)

# Greater than
result_greater_than = (15 > 10)
print("Greater Than:", result_greater_than)

# Less than
result_less_than = (8 < 12)
print("Less Than:", result_less_than)

# Greater than or equal to
result_greater_equal = (20 >= 20)
print("Greater Than or Equal To:", result_greater_equal)

# Less than or equal to
result_less_equal = (25 <= 30)
print("Less Than or Equal To:", result_less_equal)
```

# Assignment Operators:

```python
# Assignment
x = 10
print("x:", x)

# Addition Assignment
x += 5  # equivalent to x = x + 5
print("Updated x:", x)

# Multiplication Assignment
x *= 2  # equivalent to x = x * 2
print("Updated x:", x)

# Exponentiation Assignment
x **= 3  # equivalent to x = x ** 3

print("Final x:", x)
```

# Identity Operators:

Identity operators are used to check if two variables refer to the same object in memory.

    is Operator:
        Returns True if both variables refer to the same object.
    is not Operator:
        Returns True if both variables do not refer to the same object.

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

# is operator
result_is = (a is b)
print("a is b:", result_is)  # True, as a and b refer to the same list object

# is not operator
result_is_not = (a is not c)
print("a is not c:", result_is_not)  # True, as a and c do not refer to the same list object
```

# Membership Operators:

Membership operators are used to test whether a value or variable is found in a sequence (like strings, lists, or tuples).

    in Operator:
        Returns True if a value is found in the sequence.
    not in Operator:
        Returns True if a value is not found in the sequence.

```python
my_list = [1, 2, 3, 4, 5]

# in operator
result_in = (3 in my_list)
print("3 in my_list:", result_in)  # True, as 3 is present in the list

# not in operator
result_not_in = (6 not in my_list)
print("6 not in my_list:", result_not_in)  # True, as 6 is not present in the list
```

# while loop


```python
# while loop
counter = 0
while counter < 5:
    print("Counter:", counter)
    counter += 1
```

# while loop with continue and break


```python
# while loop with continue and break
num = 0
while num < 10:
    num += 1
    if num % 2 == 0:
        continue  # Skip even numbers
    print("Odd number:", num)
    if num == 7:
        break  # Exit the loop when num reaches 7
```

# for loop with continue


```python
# for loop with continue
numbers = [1, 2, 3, 4, 5]
for num in numbers:
    if num % 2 == 0:
        continue  # Skip even numbers
    print("Odd number:", num)
```

# nested for loops


```python
# nested for loops

fruits = ["apple", "banana", "cherry"]
colors = ["red", "blue", "green"]

for fruit in fruits:
    for color in colors:
        print(fruit,color)
```

# Overview of NumPy for Numerical Computing:

# Numpy Examples

```python
import numpy as np

# 1. Creating a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print("1. NumPy Array:")
print(arr)

# 2. Creating an array with regularly spaced values
arr_range = np.arange(0, 10, 2)
print("\n2. Arange:")
print(arr_range)

# 3. Creating an array with evenly spaced values over a specified range
arr_linspace = np.linspace(0, 1, 5)
print("\n3. Linspace:")
print(arr_linspace)

# 4. Creating an array filled with zeros
arr_zeros = np.zeros((3, 4))
print("\n4. Zeros:")
print(arr_zeros)

# 5. Creating an array filled with ones
arr_ones = np.ones((2, 3))
print("\n5. Ones:")
print(arr_ones)

# 6. Generating random values from a uniform distribution
rand_arr = np.random.rand(3, 2)
print("\n6. Random Values:")
print(rand_arr)

# 7. Calculating the mean of an array
mean_value = np.mean(arr)
print("\n7. Mean:")
print(mean_value)

# 8. Calculating the sum of array elements
sum_value = np.sum(arr)
print("\n8. Sum:")
print(sum_value)

# 9. Reshaping an array
arr_reshape = arr.reshape((5, 1))
print("\n9. Reshape:")
print(arr_reshape)

# 10. Calculating the dot product of two arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
dot_product = np.dot(arr1, arr2)
print("\n10. Dot Product:")
print(dot_product)
```

# Introduction to Pandas for Data Manipulation and Analysis:

Pandas:

    Pandas is a data manipulation and analysis library for Python.
    It provides data structures like Series and DataFrame for efficient data handling.
    Pandas is widely used for cleaning, exploring, and analyzing data.

```python
import pandas as pd

# Creating a Pandas DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 22],
        'City': ['New York', 'San Francisco', 'Los Angeles']}

df = pd.DataFrame(data)

# Displaying the DataFrame
print("Pandas DataFrame:")
print(df)

# Performing operations on the DataFrame
average_age = df['Age'].mean()
youngest_person = df['Name'][df['Age'].idxmin()]

print("\nAverage Age:", average_age)
print("Youngest Person:", youngest_person)
```

```python
import numpy as np

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Element-wise addition
result_add = arr1 + arr2

# Element-wise multiplication
result_multiply = arr1 * arr2

print("NumPy Array Operations:")
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Addition:", result_add)
print("Multiplication:", result_multiply)
```

```python
import pandas as pd

data1 = {'A': [1, 2, 3],
         'B': [4, 5, 6]}

data2 = {'A': [7, 8, 9],
         'B': [10, 11, 12]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# Concatenating DataFrames vertically
result_concat = pd.concat([df1, df2])

# Summing values along columns
result_sum = result_concat.sum()

print("\nPandas DataFrame Operations:")
print("DataFrame 1:")
print(df1)
print("DataFrame 2:")
print(df2)
print("Concatenation:")
print(result_concat)
print("Sum along columns:")
print(result_sum)
```

# Functions

In Python, a function is a reusable block of code that performs a specific task. Functions are defined using the def keyword, and they can take input parameters, perform operations, and return a result. Functions help in organizing code, making it modular, and allowing code reuse.

```python
def greet(name):
    """This function greets the person passed in as a parameter."""
    print("Hello, " + name + "!")

# Call the function
greet("Alice")
```

```python
# 1. Simple Function
def add_numbers(a, b):
    """Adds two numbers and returns the result."""
    return a + b

result = add_numbers(3, 7)
print("1. Sum:", result)

# 2. Function with Default Parameter
def calculate_rectangle_area(length, width=5):
    """Calculates the area of a rectangle with optional default values."""
    return length * width

area1 = calculate_rectangle_area(8)
area2 = calculate_rectangle_area(8, 6)
print("\n2. Area 1:", area1)
print("   Area 2:", area2)

# 3. Function with Variable Number of Arguments
def calculate_average(*numbers):
    """Calculates the average of a variable number of arguments."""
    return sum(numbers) / len(numbers)

avg1 = calculate_average(2, 4, 6, 8)
avg2 = calculate_average(1, 3, 5, 7, 9)
print("\n3. Average 1:", avg1)
print("   Average 2:", avg2)

# 4. Function Returning Multiple Values
def calculate_square_and_cube(number):
    """Returns the square and cube of a number."""
    square = number ** 2
    cube = number ** 3
    return square, cube

square_result, cube_result = calculate_square_and_cube(4)
print("\n4. Square:", square_result)
print("   Cube:", cube_result)

# 5. Recursive Function
def factorial(n):
    """Calculates the factorial of a number using recursion."""
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

result_factorial_5 = factorial(5)
result_factorial_8 = factorial(8)
print("\n5. Factorial of 5:", result_factorial_5)
print("   Factorial of 8:", result_factorial_8)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)
plt.figure(figsize=(10, 8))

#plotting
plt.plot(t, s,color='r',linestyle='--',linewidth=10)

plt.xlabel('time (s)') 
plt.ylabel('voltage (mV)')
plt.title('About as simple as it gets, folks')
plt.grid()
plt.show()
```
